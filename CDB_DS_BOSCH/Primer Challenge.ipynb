##### **Lectura del Archivo de Texto en un DataFrame**

```python
column_widths = [4, 3, 3, 9, 5, 6, 5,2] #Espacio entre Columnas
column_names = ["Año", "Mes", "Día", "Año_Fraccionado", "Mancha_Solar", "Desviación_Estandar", "Número_de_Observaciones", "Indicador_Definitivo/Provisional"] #Nombre de las columnas
df = pd.read_fwf(Data, widths = column_widths, names = column_names)  #Leer Archivo con nuevo formato

print(df) #Muestra el DataFrame Resultante
```

##### **Copia de Datos y Modificaciones en la Copia** #####

```python
data = df.copy()
data["Año"] = data["Año"].astype(str) + "_" + data["Mes"].astype(str) + "_" + data["Día"].astype(str) #Unión de Columnas
data = data.drop(["Mes", "Día"], axis= "columns") #Eliminando Columnas Duplicadas
data = data.rename(columns={"Año" : "Año_Mes_Día"}) #Renombrar Columna

data.head() #Mostrar
```

##### **Graficación Rápida**

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Año_Fraccionado'], df['Mancha_Solar'], marker = 'o', linestyle = '-', color = 'm')
plt.xlabel('Ano Fraccionado')
plt.ylabel('Mancha Solar') 
plt.title('Evolución de Manchas Solares a lo largo del Tiempo')
plt.grid(True)
plt.show()
```

Evolución de las manchas solares a lo largo de los años, se puede apreciar que no tienen un crecimiento fijo, mas bien una tendencia a ser estacionales.

```python
plt.figure(figsize = (10, 6))
plt.scatter(df['Mancha_Solar'], df['Desviación_Estandar'], color = 'm', alpha = 0.5)
plt.xlabel('Manchas Solares') 
plt.ylabel('Desviación Estándar')
plt.title('Relación entre Manchas Solares y Desviación Estándar') 
plt.grid(True)
plt.show()
```

Relación de la **Desviación Estandar** con las **Manchas Solares**, dado que sigue una tendencia entonces existe una relación. Hay un creciimiento con datos atípicos.

```python
# Nombre del archivo ZIP
nombre_archivo_zip = 'diary14.zip'

# Ruta completa al archivo ZIP
ruta_zip = os.path.join(ruta_carpeta, nombre_archivo_zip)

# Carpeta temporal para extraer los archivos del ZIP
carpeta_temporal = 'Data'

# Extraer el archivo ZIP en la carpeta temporal
with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
    zip_ref.extractall(carpeta_temporal)

# Lista para almacenar los DataFrames de los archivos CSV
dataframes = []

# Recorre todos los archivos en la carpeta temporal
for archivo in os.listdir('./Data/diary14/'):
    if archivo.endswith('.csv'):
        ruta_csv = os.path.join('./Data/diary14/', archivo)
        df = pd.read_csv(ruta_csv)
        dataframes.append(df)
```

##### **Contabilizando los DataFrames**

```python
num_dataframes = len(dataframes)
print(f"La carpeta contiene {num_dataframes} DataFrames.")
```

##### **Contabilizando las Columnas de cada DataFrame**

```python
for idx, df in enumerate(dataframes):
    num_columnas = df.shape[1]
    print(f"DataFrame {idx} tiene {num_columnas} columnas.")
```
